# Machine learning

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df,train_size=0.8,random_state=42,stratify=df['smoking'])
```

```python
from catboost import CatBoostClassifier
```

```python
y  = ['smoking']
X = list(df.drop(columns = ['smoking']).columns)
cat_features = list(df.select_dtypes(include = ['object']).columns)
```

```python
from catboost import Pool

train_data = Pool(data=train[X],
                  label=train[y],
                  cat_features=cat_features
                 )

test_data = Pool(data=test[X],
                  label=test[y],
                  cat_features=cat_features
                 )
```

```python
params = {'verbose':100,
          'eval_metric':'F1',
          'loss_function': 'Logloss',
          'random_seed':42,
          'learning_rate':0.02,
         'auto_class_weights':'Balanced'}
```

```python
Cat_model = CatBoostClassifier(**params)
```

```python
from catboost import cv
```

```python
%time
cv_data = cv(
    params = params,
    pool = train_data,
    fold_count=5,
    shuffle=True,
    partition_random_seed=0,
    stratified=False,
    verbose=False
)
```

```python
n_iters = cv_data[cv_data['test-F1-mean'] == cv_data['test-F1-mean'].max()]['iterations'].values[0]
```

```python
n_iters
```

```python
params = {'iterations':n_iters + 1,
          'verbose':100,
          'eval_metric':'F1',
          'loss_function': 'Logloss',
          'random_seed':42,
          'learning_rate':0.01,
         'auto_class_weights':'Balanced'}
```

```python
Cat_model = CatBoostClassifier(**params)
model = Cat_model.fit(train_data)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, f1_score
```

```python
probabilities = (model.predict_proba(test_data)[:,1])
score = model.predict(test_data)
fpr, tpr, thr = roc_curve(test['smoking'], probabilities)
auc_roc = roc_auc_score(test['smoking'], probabilities)
f1 = f1_score(test['smoking'],score)
```
