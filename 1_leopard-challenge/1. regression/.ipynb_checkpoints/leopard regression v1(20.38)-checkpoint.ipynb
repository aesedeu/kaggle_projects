# Ð’Ñ‹Ð¿Ð¾Ð»Ð½ÑÐµÐ¼ Ð¿Ñ€ÐµÐ´Ð¾Ð±Ñ€Ð°Ð±Ð¾Ñ‚ÐºÑƒ Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð¸ Ð¸Ð·Ð±Ð°Ð²Ð»ÑÐµÐ¼ÑÑ Ð¾Ñ‚ NaN

```python
df.isna().mean()
```

```python
sp = pd.Series(df.columns).tolist()
```

```python
sp
```

```python
for i in sp:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.isna().mean()
```

```python
df.shape
```

# ÐÐ°Ñ‡Ð¸Ð½Ð°ÐµÐ¼ ÑÐ±Ð¾Ñ€ÐºÑƒ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð¼Ð°ÑˆÐ¸Ð½Ð½Ð¾Ð³Ð¾ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ñ

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df,train_size=0.6,random_state=42)
val, test = train_test_split(test,train_size=0.5,random_state=42)
```

```python
train.columns
```

```python
train.select_dtypes(include='object')
```

```python
X = ['Suburb', 'Rooms', 'Type', 'Method',
     'Distance', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 
     'CouncilArea', 'Regionname', 'Propertycount']

cat_features = ['Suburb','Type','Method','CouncilArea','Regionname']

y = ['Price']
```

```python
from catboost import Pool

train_data = Pool(data=train[X],
                  label=train[y],
                  cat_features=cat_features,
                 )

valid_data = Pool(data=val[X],
                  label=val[y],
                  cat_features=cat_features
                 )
```

```python
from catboost import CatBoostRegressor
```

```python
parameters = {'cat_features': cat_features,
              'eval_metric': 'MAPE',
              'loss_function':'RMSE',
              'random_seed':42,
              'learning_rate':0.1,
              'verbose':100}
```

# ÐšÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ñ Ð¿Ð¾Ð´Ð±Ð¾Ñ€Ð° Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð²

```python
from catboost import cv
```

```python
cv_data = cv(
    params = parameters,
    pool = train_data,
    fold_count=5,
    shuffle=True,
    partition_random_seed=0,
    stratified=False,
    verbose=False,
    early_stopping_rounds=200
)
```

```python
cv_data.head()
```

```python
cv_data[cv_data['test-MAPE-mean'] == cv_data['test-MAPE-mean'].min()]
```

```python
model = CatBoostRegressor(**parameters)
```

```python
model.fit(train_data, eval_set=valid_data)
```

# Ð”Ð¾Ð±Ð°Ð²Ð»ÑÐµÐ¼ Ñ‚ÐµÑÑ‚Ð¾Ð²Ñ‹Ð¹ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

def error(y_true,y_pred):
  print(mean_absolute_error(y_true,y_pred))
  print(mean_absolute_percentage_error(y_true,y_pred))
```

```python
test['price_pred_all'] = model.predict(test[X])
```

```python
error(test['Price'],test['price_pred_all'])
```
