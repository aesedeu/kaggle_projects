```python
df = df.replace(np.nan, 0)
```

```python
df['year'] = df["year"].astype(np.int64)
```

```python
df = df.replace(0, np.nan)
```

```python
df_year_price = df.groupby('year')['price'].agg('mean').reset_index()
df_year_price = pd.DataFrame(df_year_price)
df_year_price
```

```python
df_year_price['group'] = pd.qcut(df_year_price['year'], 8)
```

```python
df_year_price.groupby('group')['price'].agg(['count','median','mean', 'max']).reset_index()
```
